```python
from PIL import Image
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import numpy as np
```

```python
img = Image.open("tiger.png")
```

```python
img_np = np.asarray(img)
```

```python
#print(img_np)
```

```python
img_np.shape
```

```python
pixels = img_np.reshape(img_np.shape[0]*img_np.shape[1], img_np.shape[2])
```

```python
pixels.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=16)
```

```python
model.fit(pixels)
```

```python
pixel_cen = model.labels_
```

```python
cluster_centers = model.cluster_centers_
```

```python
pixel_cen[0]
```

```python
final = np.zeros((pixel_cen.shape[0],3))
```

```python
for i in range(16):
    final[pixel_cen==i]=cluster_centers[i]
    
```

```python
final[0:10]
```

```python
## rehsape
```

```python
comp = final.reshape(img_np.shape[0], img_np.shape[1],3)
```

```python
comp.shape
```

```python
comp_img = Image.fromarray(np.uint8(comp))
```

```python
comp_img.save("tiger_comp.png")
```
